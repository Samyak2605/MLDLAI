```python
import gensim
import os
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
file_names = [
    'avengers-endgame-script-slug.txt',
    'avengers-script-slug.txt',
    'black-panther-script-slug.txt',
    'captain_america.txt',
    'captain_marvel.txt',
    'civil_war.txt',
    'guardians_2.txt',
    'infinity_war.txt',
    'iron_man_2.txt',
    'iron_man_3.txt',
    'iron-man-script-slug.txt',
    'spider_man_homecoming.txt',
    'thor_dark_world.txt',
    'thor-ragnarok-script-slug.txt',
    'thor-script-slug.txt',
    'winter_soldier.txt'
]
for filename in file_names:

    f = open(filename)
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('clint')
```
